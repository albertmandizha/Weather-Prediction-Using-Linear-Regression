# Linear Regression Model on the Weather Data

```python
#Pandas Makes reading files and input files quick
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn import preprocessing
import numpy as np
from sklearn.linear_model import LinearRegression
#from mixtend.evaluate import bias_viariance_decomp
import datetime

data =pd.read_csv('weatherHistory.csv')
data
```

```python
data["Date"]=pd.to_datetime(data["Formatted Date"],utc=True)
data["Month"]= data["Date"].dt.month
data
```

```python
#deleting unnecessary columns
data=data.drop(['Formatted Date','Daily Summary','Pressure (millibars)','Loud Cover','Date'],axis=1)
data
```

```python
label_encoder=preprocessing.LabelEncoder()
data['Summary']=label_encoder.fit_transform(data['Summary'])
data['Summary'].unique()
data
```

```python
data['Precip Type']=data["Precip Type"].replace(['rain','snow'],[0,1])
data
```

```python
#checking for NaN values in data
data.isna().sum()
```

```python
#replacing NaN values with 0
data['Precip Type']=data['Precip Type'].replace(np.nan,0)
data.isna().sum()
```

```python
y=data['Apparent Temperature (C)']
data=data.drop('Apparent Temperature (C)',axis = 1)
print(data)
```

```python
y
```

```python
x_train,x_test,y_train,y_test=train_test_split(data,y,test_size=0.2,random_state=0)

model=LinearRegression()
model.fit(x_train,y_train)
```

```python
pred=model.predict(x_test)

#finding variance using np
variance=np.var(pred)
variance
```

```python
SSE=np.mean((np.mean(pred)-y)**2)
Bias=SSE-variance
Bias
```
